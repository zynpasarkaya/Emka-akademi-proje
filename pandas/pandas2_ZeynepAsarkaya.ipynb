7) 6.Soruda erkeklerin ve kadınların ödedikleri bahşişler toplamına bakıldığında Erkeklerin kadınlara kıyasla toplamda daha çok bahşiş ödediğini görüyoruz. Fakat biz biliyoruz ki bizim verimizde 157 Erkek, 87 kadın verisi vardı. Dolayısıyla Erkek sayısı daha fazla olduğu için toplam ödeyecekleri bahşiş miktarlarında da daha fazla çıkması oldukça normal bir durum. Bu yüzden daha mantıklı bir çıkarım için bu sefer df_male ile df_female'in bahşiş ORTALAMALARIN'a bakıp kıyaslayın, aralarında anlamlı ve ciddi bir fark var mı ?

```python
dfmaleort = df_male['tip'].mean()
dffemaleort = df_female['tip'].mean()
print(dfmaleort)
print(dffemaleort)
```

 Oluşturmuş olduğunuz df_male ile çalışarak elimizdeki verideki erkeklerin yüzde kaçının sigara içtiğini bulun. Aynı şekilde elimizdeki kadınların yüzde kaçının sigara içtiğini de bulun. Aralarında anlamlı bir fark mı yorumlayın.

```python
dfmaley = len(df_male[df_male['smoker'] == 'Yes'])
dfmaletop = len(df_male['smoker'])
dfmaleson = (dfmaley / dfmaletop) * 100

dffemaley = (df_female['smoker'] == 'Yes').sum()
dffemaletop = len(df_female['smoker'])
dffemaleson = (dffemaley / dffemaletop) * 100
print(dfmaleson)
print(dffemaleson)
```

9) Erkeklerin zaman olarak dinnerda mı yoksa lunch'da mı daha fazla yemek yediğini yüzdesel olarak bulun. Aynı işlemi kadınlar için de yapın. Yüzdesel olarak erkeklerin ve kadınların yemek yeme vakitleri arasında anlamlı bir fark var mı yok mu yorumlayın.

```python
maledin = len(df_male[df_male['time'] == 'Dinner'])
malelun = len(df_male[df_male['time'] == 'Lunch'])
maletop = len(df_male['time'])

malediny = (maledin / maletop) * 100
maleluny = (malelun / maletop) * 100
print(malediny,maleluny)

femaledin = len(df_female[df_female['time'] == 'Dinner'])
femalelun = len(df_female[df_female['time'] == 'Lunch'])
femaletop = len(df_female['time'])

femalediny = (femaledin / femaletop) * 100
femaleluny = (femalelun / femaletop) * 100
print(femalediny,femaleluny)
```
