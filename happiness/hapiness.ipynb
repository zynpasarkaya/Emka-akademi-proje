```python
df = df.groupby(['Country or region']).mean(numeric_only=True).sort_values('Score',ascending=False) 
df
```

```python
# En mutlu 10 ülke

df = df.groupby(['Country or region']).mean(numeric_only=True).sort_values('Score',ascending=False) 
df.head(10)
```

```python
#En mutsuz 10 ülke
df = df.groupby(['Country or region']).mean(numeric_only=True).sort_values('Score',ascending=False) 
df.tail(10)
```

```python
ax = df.plot( y=['GDP per capita','Social support', 'Healthy life expectancy','Freedom to make life choices','Perceptions of corruption','Generosity'],figsize=(15,5), kind='bar', rot=45)
ax.set_ylabel('Score')
ax.set_title('')
plt.show()
```

```python
'''Korelasyon, iki değişken arasındaki ilişkinin gücünü ve yönünü ölçer. 
   Pozitif bir korelasyon, değişkenler arasında doğru orantılı bir ilişki olduğunu gösterirken, 
   negatif bir korelasyon ters orantılı bir ilişki olduğunu gösterir.'''

dfcorr = df[['GDP per capita','Social support', 'Healthy life expectancy',
             'Freedom to make life choices','Perceptions of corruption','Generosity']].corr()
dfcorr
```

```python
df2 = pd.read_csv('world-happiness-report.csv')
df2
```

```python
df2 = df2.rename(columns={'Log GDP per capita':'Economy (GDP per Capita)',
                       'Life Ladder':'Happiness Score'})
df2.head()
```

```python
tur = df2[(df2['Country name'] == 'Turkey')]
dt = tur[['year','Happiness Score', 'Economy (GDP per Capita)']].copy()
fig, ax5 = plt.subplots(figsize=(16,8)) 
dt.plot(x = 'year', y = 'Happiness Score', ax = ax5) 
dt.plot(x = 'year', y = 'Economy (GDP per Capita)', ax = ax5, secondary_y = True) 
ax5.set_title('Mutluluk ve ekonomi arasındaki ilişki ( Türkiye )')
plt.show()
```
